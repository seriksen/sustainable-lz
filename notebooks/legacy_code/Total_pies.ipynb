```python
import matplotlib.patches as mpatches
values = {
    'UKDC': 40, # 23 from storage, 17 from jobs
    'USDC': 11, # 1.2 from HPSS, 10 from jobs
    'Flights': 144, # 250-320 collab meetings a year
    'Operations': 1142#*0.85
}
# plot a pie chart
colors=['peru','orange','r','b']
size=0.25
plt.pie(values.values(), labels=values.keys(),  radius=1-size,colors=colors,
       wedgeprops=dict(width=size, edgecolor='w'))

values = {
    'UKDC': 40, # 23 from storage, 17 from jobs
    'USDC': 11, # 1.2 from HPSS, 10 from jobs
    'Travel': 144, # 250-320 collab meetings a year
    'Operations': 1142#*0.85
}
plt.pie(values.values(), radius=1-2*size,colors=colors,
       wedgeprops=dict(width=size, edgecolor='w'),hatch='o')


patches=[]
labels=[]
patches.append(
    ( mpatches.Patch(facecolor='r',edgecolor='white',hatch='o'),
    )
)
labels.append('Median (50%CL)')

patches.append(
    ( mpatches.Patch(facecolor='r'),
    )
)
labels.append('Conservative (90%CL)')

leg=plt.legend(patches,labels,frameon=False, bbox_to_anchor=[0.7,0.79],handlelength=3,title='Flight non-CO2e factor')
for patch in leg.get_patches():
    patch.set_height(12)
    patch.set_y(-6)
# label=
# plt.show()
plt.savefig("total_plot.png",bbox_inches="tight")
```
