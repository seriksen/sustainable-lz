```python
import matplotlib.patches as mpatches
import numpy as np
values = {
    'UKDC': nukdc, # 23 from storage, 17 from jobs
    'USDC': nusdc, # 1.2 from HPSS, 10 from jobs
    'Flights': flights_low_M*3/1.7, # 250-320 collab meetings a year
    'Operations': nOps#*0.85
}
# plot a pie chart
colors=['peru','orange','r','b']
size=0.25
plt.pie(values.values(), labels=values.keys(),  radius=1-size,colors=colors,
       wedgeprops=dict(width=size, edgecolor='w'))

values = {
    'UKDC': nukdc, # 23 from storage, 17 from jobs
    'USDC': nusdc, # 1.2 from HPSS, 10 from jobs
    'Travel': flights_low_M, # 250-320 collab meetings a year
    'Operations': nOps#*0.85
}
plt.pie(values.values(), radius=1-2*size,colors=colors,
       wedgeprops=dict(width=size, edgecolor='w'),hatch='o')


patches=[]
labels=[]
patches.append(
    ( mpatches.Patch(facecolor='r',edgecolor='white',hatch='o'),
    )
)
labels.append('100 year')

patches.append(
    ( mpatches.Patch(facecolor='r'),
    )
)
labels.append('30 year')

leg=plt.legend(patches,labels,frameon=False, bbox_to_anchor=[0.75,0.79],handlelength=3,title='Non-CO2e Time Horizon')
for patch in leg.get_patches():
    patch.set_height(12)
    patch.set_y(-6)
# label=
# plt.show()
plt.title("Total: %.f (%.f) tCO2e/yr for 30(100) year time horizon"%(np.round(ntotal+flights_low_M*3/1.7,-1),np.round(ntotal+flights_low_M,-1)))
plt.savefig("total_high_assumption.png",bbox_inches="tight")
```
