```python
values = {
    'UKDC': {'storage': 23, 'jobs': 17},
    'USDC': {'storage': 24, 'jobs': 10},
    'Travel': {'low': 250, 'high': 320},
    'Operations': {'low': 375, 'high': 1142}
}
# Extract keys and data
categories = list(values.keys())
stacked_categories = ['UKDC', 'USDC']
error_categories = ['Travel', 'Operations']

# Prepare data for stacked bar
ukdc_data = [values['UKDC']['storage'], values['UKDC']['jobs']]
usdc_data = [values['USDC']['storage'], values['USDC']['jobs']]

# Prepare data for error bars
travel_mid = (values['Travel']['low'] + values['Travel']['high']) / 2
travel_error = np.array([travel_mid - values['Travel']['low'], values['Travel']['high'] - travel_mid])

operations_mid = (values['Operations']['low'] + values['Operations']['high']) / 2
operations_error = np.array([operations_mid - values['Operations']['low'], values['Operations']['high'] - operations_mid])

# Plotting
fig, ax = plt.subplots()

# Stacked bars for UKDC and USDC
ax.bar('UKDC', ukdc_data[0], color='red', label=f'UKDC: {ukdc_data[0] + ukdc_data[1]}' + r' $tCO_2$/year')
ax.bar('UKDC', ukdc_data[1], bottom=ukdc_data[0], hatch='//')

ax.bar('USDC', usdc_data[0], color='blue', label=f'USDC: {usdc_data[0] + usdc_data[1]}' + r' $tCO_2$/year')
ax.bar('USDC', usdc_data[1], bottom=usdc_data[0], hatch='//')

# Error bars for Travel and Operations
ax.bar('Travel', travel_mid, yerr=[[travel_error[0]], [travel_error[1]]], capsize=5, color='lightblue', label=f'Travel: {travel_mid}' + r' $tCO_2$/year')
ax.bar('Operations', operations_mid, yerr=[[operations_error[0]], [operations_error[1]]], capsize=5, color='lightgreen', label=f'Operations: {operations_mid}' + r' $tCO_2$/year')

# Labeling and display
ax.set_ylabel(r'$tCO_2$ / year')
ax.legend()

plt.show()
```

```python
# Data for pie chart
ukdc_total = sum(values['UKDC'].values())
usdc_total = sum(values['USDC'].values())
travel_mid = (values['Travel']['high'])
operations_mid = (values['Operations']['high'])

# Pie chart values
pie_values = [ukdc_total, usdc_total, travel_mid, operations_mid]
labels = ['UKDC', 'USDC', 'Travel', 'Operations']

# Plotting pie chart
fig, ax = plt.subplots()
ax.pie(pie_values, labels=labels)

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')  

plt.show()
```
